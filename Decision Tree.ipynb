```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing the Data

```python
df=pd.read_csv('/home/sai/Downloads/Iris.csv')
df=df.iloc[:,1:]
df.head()
```

### Preparing the train and test datasets

```python
X=df.iloc[:,:4].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0) 
```

### Defining the Decision Tree Classifier Model and fitting on train data

```python
from sklearn.tree import DecisionTreeClassifier

dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)

print('Decision Tree Classifer Created')
```

### Predictions on test data and accuracy score

```python
dtree.predict(X_test)
```

```python
y_test
```

```python
dtree.score(X_test,y_test)
```

This Decision Tree model is able to classify the data with an accuracy of 96.67%

### Visualizing the Decision Tree Classifier Graphically

```python
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=df.iloc[:,:4].columns,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
